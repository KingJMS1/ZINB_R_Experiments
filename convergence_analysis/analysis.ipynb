```r
avg_obs_per_spatial_point <- c(2, 4, 8, 16, 24, 32, 48, 64)
```

```r
toPlot <- list(l1tcover, l2tcover, sigma1tcover, sigma2tcover, sigma1scover, sigma2scover, l1scover, l2scover, snoisecover, tnoisecover, alphacover, betacover, acover, bcover, ccover, dcover)
names <- c("l1t", "l2t", "sigma1t", "sigma2t", "sigma1s", "sigma2s", "l1s", "l2s", "snoise", "tnoise", "alpha", "beta", "a", "b", "c", "d")

covers <- matrix(0, nrow=length(avg_obs_per_spatial_point), ncol=length(names))

i <- 1
for(var in toPlot)
{
    png(paste("../plots/conv", names[i], ".png", sep=""), width = 800, height = 800, units = "px", res = 150) 
    plot(avg_obs_per_spatial_point, colMeans(var), type="o", main=paste(names[i], "coverage of 95% interval"))
    covers[,i] <- colMeans(var)
    dev.off()
    i <- i + 1
}
```

```r
pretty <- as.data.frame(covers, row.names = avg_obs_per_spatial_point)
colnames(pretty) <- names
```
