```python
def kern(a, b):
    return np.exp(-np.square(a[:, None] - b[None, :]))
```

```python
def b_i(i, c):
    if i == c.shape[0]:
        return None
    csn = c[i, (i + 1):]
    cn = c[(i + 1):, (i + 1):]
    return np.linalg.solve(cn, csn) 

def b_st_i(i, c):
    shp = c.shape[0]
    b_st = np.ones(shp)
    mask = np.ones(shp, dtype=bool)
    mask[i] = False
    b = b_i(i, c)
    b_st[mask] = -b
    return b_st

def f_i(i, c):
    csn = np.delete(c[i], i)
    cn = np.delete(c, i, 0)
    cn = np.delete(cn, i, 1)
    return 1 - csn @ np.linalg.solve(cn, csn)

f = np.array([f_i(i, c) for i in range(c.shape[0])])
b_st = np.array([b_st_i(i, c) for i in range(c.shape[0])])
```
